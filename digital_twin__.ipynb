### CORRELATION PLOTS

#### Response percentage and Actual Value

```python
import numpy as np

# Calculate the line of best fit
m, b = np.polyfit(q['response_perc_q'], q['ActualValue'], 1)

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(q['response_perc_q'], q['ActualValue'], color='black',alpha=0.5, s=10, label='Data points')
plt.plot(q['response_perc_q'], m * q['response_perc_q'] + b, color='red', label='Best fit line')

# Adding title and labels
plt.title('Correlation between response_perc_q and ActualValue')
plt.xlabel('response_perc_q')
plt.ylabel('ActualValue')
plt.legend()
plt.grid(True)
```

### DATA PREPARATION

```python
filtered_df = q[q['Question'].str.startswith('Scale') | q['Question'].eq('Q26') | q['Question'].eq('Q27')]
```

```python
filtered_df.head()
```

```python
def transform_df(df):
    # Pivot the DataFrame to get ActualValue for each Scale and Q27
    pivot_df = df.pivot_table(index=['UKPRN', 'Subject Code'],
                              columns='Question',
                              values='ActualValue').reset_index()

    return pivot_df
```
